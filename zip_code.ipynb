# Appending demographic information to dataset using ZIP codes

```python
from uszipcode import SearchEngine, SimpleZipcode, ComprehensiveZipcode
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\email\OneDrive\Documents\Python\zip_codes\cluster_data_pregrouping.csv")
df.head(10)
```

```python
print(df.shape)
```

```python
se = SearchEngine()
zip_df = []
for i in df["ZipCode"].dropna().unique():
    row = [i]
    x = se.by_zipcode(i)
    row.append(x.lat)
    row.append(x.lng)
    row.append(x.bounds_north)
    row.append(x.bounds_south)
    row.append(x.bounds_east)
    row.append(x.bounds_west)
    row.append(x.radius_in_miles)
    row.append(x.land_area_in_sqmi)
    row.append(x.water_area_in_sqmi)

    row.append(x.population)
    row.append(x.population_density)

    row.append(x.housing_units)
    row.append(x.occupied_housing_units)
    row.append(x.median_home_value)
    row.append(x.median_household_income)
    zip_df.append(row)

zipped = pd.DataFrame(zip_df, columns=["ZipCode", "Latitude", "Longitude", "BoundNorth", "BoundSouth", "BoundEast", "BoundWest", "RadiusInMiles", "LandArea", "WaterArea", "Population", "PopulationDensity", "HousingUnits", "OccupiedHousingUnits", "MedianHomeValue", "MedianHouseholdIncome" ])
zipped.head(10)

    
```

```python
# zipped.to_csv('zipped.csv')
print(len(zipped["ZipCode"]))
print(len(zipped["ZipCode"].unique()))
```

```python
output = pd.merge(left = df, right = zipped, how =  'left', on = 'ZipCode')
new_output = []
for i in output["CCN"]:
    i = str(i)
    while(len(i) < 6):
        i = "0" + i
    new_output.append(i)
output['CCN'] = new_output
output.head(10)
```

```python
output.to_csv("output.csv")
output.shape
```
